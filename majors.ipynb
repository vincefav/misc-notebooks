```python
recentgrads['Expected_value'] = recentgrads['Median'] * (1-recentgrads['Unemployment_rate'])
gradstudents['Expected_value'] = gradstudents['Grad_median'] * (1-gradstudents['Grad_unemployment_rate'])
```

```python
value_of_majors = recentgrads.sort_values(by='Expected_value',ascending=True)
value_of_majors.set_index('Major',inplace=True)
```
